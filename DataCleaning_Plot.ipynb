```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\Users\\Athiban\\Desktop\\Kee\\movies_metadata.csv')
budget_cl = data.budget
bud_clm = pd.to_numeric(budget_cl, errors='coerce')

fig = plt.figure(figsize = (10, 5))
profit = data['revenue'] - bud_clm / bud_clm
plt.plot(profit)
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('profit check')
plt.show()  
```

```python
import ast
genres = set([])

for i in data.genres:
  genList = ast.literal_eval(i)
  for j in genList:
    genres.add(j['name'])
```

```python
rows = [[]]
genres = list(genres)

for i in data.genres:
  genList = ast.literal_eval(i)
  movieGenres = [x['name'] for x in genList] # list of all the genres of the movie
  row = [] # store a list of 1s and 0s 
  for i in genres:
    row.append(int(i in movieGenres))
  rows.append(row)
```

```python
genre = pd.DataFrame(rows[1:],columns = genres)
genre
```
